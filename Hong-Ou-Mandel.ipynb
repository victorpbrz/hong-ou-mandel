## Hong-Ou-Mandel

## unitária: $\ \ U_{DF}=\overset{}{\underset{}\exp \big\{i\theta\big(b^\dagger a+ba^\dagger\big)}\big\}$

## p/ implementação IBMq $\to$ operadores em termos das matrizes de Pauli

## $\begin{align}b_P^\dagger&=\sum_{n=0}^N\sqrt{n+1}\ \ b_{n\to n+1}^\dagger\\[5pt] b_P&=\sum_{n=0}^N\sqrt{n}\ \ b_{n\to n-1}\end{align}$

## $\begin{array}{}b_P^\dagger= b^\dagger_{0\to 1}+\sqrt2\ b^\dagger_{1\to 2}+\sqrt3\ b^\dagger_{2\to 3} + \cdots  \\[10pt] b_P=b_{1\to 0}+\sqrt2\ b_{2\to 1}+\sqrt3\ b_{3\to 2}+\cdots\end{array}$

##
## simplificações:

## $i.$ truncamento do espaço de Hilbert
## $\begin{array}{}b_P^\dagger= b^\dagger_{0\to 1}+\sqrt2\ b^\dagger_{1\to 2}+\sqrt3\ b^\dagger_{2\to 3},  \\[10pt] b_P=b_{1\to 0}+\sqrt2\ b_{2\to 1}+\sqrt3\ b_{3\to 2}.\end{array}$

## $ii.$ remoção de termos que envolvem $|3\rangle_F$

## $\begin{array}{}b_P^\dagger= b^\dagger_{0\to 1}+\sqrt2\ b^\dagger_{1\to 2}, \\[10pt] b_P=b_{1\to 0}+\sqrt2\ b_{2\to 1}.\end{array}$

## $iii.$ remoção de termos que não envolvem 2 fótons no total

### $\begin{align}b^\dagger_P a_P&=\big(b^\dagger_{0\to 1}+\sqrt2\ b^\dagger_{1\to 2}\big)\big(a_{1\to 0}+\sqrt2\ a_{2\to 1} \big) \\[5pt] &= b^\dagger_{0\to 1}a_{1\to 0} +  \sqrt2\ b^\dagger_{0\to 1}a_{2\to 1} + \sqrt2\ b^\dagger_{1\to 2}a_{1\to 0} + 2\ b^\dagger_{1\to 2}a_{2\to 1} \\[5pt]&\approx \sqrt2\ \big(b^\dagger_{0\to 1}a_{2\to 1}+ b^\dagger_{1\to 2}a_{1\to 0}\big). \\[10pt]b_P a_P^\dagger&= \big(b_{1\to 0}+\sqrt2\ b_{2\to 1} \big)\big(a^\dagger_{0\to 1}+\sqrt2\ a^\dagger_{1\to 2}\big) \\[5pt]&=b_{1\to 0}a^\dagger_{0\to 1}+\sqrt2\ b_{1\to 0}a^\dagger_{1\to 2} + \sqrt2\ b_{2\to 1}a^\dagger_{0\to 1} + 2\ b_{2\to 1}a^\dagger_{1\to 2}\\[5pt]&\approx \sqrt2\ \big(b_{1\to 0}a^\dagger_{1\to 2}+ b_{2\to 1}a^\dagger_{0\to 1}\big).\end{align}$

## $b^\dagger_Pa_P + b_Pa^\dagger_P \approx\sqrt2\big\{\big(b^\dagger_{0\to 1}a_{2\to 1}+h.c.)+ (b^\dagger_{1\to 2}a_{1\to 0} + h.c.\big)\big\}$

## em termos de $\mathcal P$ e $\mathcal Q:$

### $b^\dagger_Pa_P + b_Pa^\dagger_P \approx\sqrt2\big\{\big(\mathcal P_0\otimes \mathcal Q_1\otimes \mathcal Q_0 \otimes \mathcal P_1+h.c.)+ (\mathcal Q_1 \otimes \mathcal P_1 \otimes \mathcal P_0 \otimes \mathcal Q_0  + h.c.\big)\big\}, $

## que é a eq. 16 do artigo.

## Escrevendo esse resultado em termos das matrizes de Pauli e substituindo na unitária obtemos
### $\begin{align}U_{DF}&=\exp\left(i\frac{\sqrt2\pi}{32}\big(IXXI-IXXZ+IYYI-IYYZ+ZXXI-ZXXZ+ZYYI-ZYYZ\right.\\&\left.\phantom{\frac{\sqrt2\pi}{32}}+XIIX+XIZX-XZIX-XZZX+YIIY+YIZY-YZIY-YZZY\big)\right),\end{align}$

## em que foi utilizada a notação $IXZX \equiv I^b \otimes X^b \otimes Z^a \otimes X^a$.

# $e^{i\theta H} = \lim_{j\rightarrow\infty}\Big(e^{i(\theta/j) H_{1}}e^{i(\theta/j) H_{2}}\Big)^{j}.$

## $\begin{align}j=1: \ e^{i\theta H}&\approx \big(e^{i\theta H_1}\big)\big(e^{i\theta H_2}\big)\\[5pt]&\approx \big(e^{i(\sqrt2\pi/32)(XIIX+XIZX-XZIX-XZZX+YIIY+YIZY-YZIY-YZZY \big)}\big)\\&\big(e^{i(\sqrt2\pi/32)\big(IXXI-IXXZ+IYYI-IYYZ+ZXXI-ZXXZ+ZYYI-ZYYZ\big)}\big)\\[5pt]&\approx \big(e^{i(\sqrt2\pi/32)XIIX}e^{i(\sqrt2\pi/32)XIZX}e^{i(-\sqrt2\pi/32)XZIX}e^{i(-\sqrt2\pi/32)XZZX}\\ &\ \ \ \ e^{i(\sqrt2\pi/32)YIIY}e^{i(\sqrt2\pi/32)YIZY}e^{i(-\sqrt2\pi/32)YZIY}e^{i(-\sqrt2\pi/32)YZZY}\big)\\ &\big(e^{i(\sqrt2\pi/32)IXXI}e^{i(-\sqrt2\pi/32)IXXZ}e^{i(\sqrt2\pi/32)IYYI}e^{i(-\sqrt2\pi/32)IYYZ}\\ &\ \ \ \ e^{i(\sqrt2\pi/32)ZXXI}e^{i(-\sqrt2\pi/32)ZXXZ}e^{i(\sqrt2\pi/32)ZYYI}e^{i(-\sqrt2\pi/32)ZYYZ}\big)\end{align}$

```python
# numero de passos
j = 1

# arquivo com as definições
%run PORTAS_DEF.ipynb
```

```python
#define H_1
H_1 = QuantumCircuit(4, name=r"$\exp(i\theta H_1)$")
H_1.append(YZZY(), [[0], [1], [2], [3]])
H_1.append(YZIY(), [[0], [1], [2], [3]])
H_1.append(YIZY(), [[0], [1], [2], [3]])
H_1.append(YIIY(), [[0], [1], [2], [3]])
H_1.append(XZZX(), [[0], [1], [2], [3]])
H_1.append(XZIX(), [[0], [1], [2], [3]])
H_1.append(XIZX(), [[0], [1], [2], [3]])
H_1.append(XIIX(), [[0], [1], [2], [3]])

# define H_2
H_2 = QuantumCircuit(4, name=r"$\exp(i\theta H_2)$")
H_2.append(ZYYZ(), [[0], [1], [2], [3]])
H_2.append(ZYYI(), [[0], [1], [2], [3]])
H_2.append(ZXXZ(), [[0], [1], [2], [3]])
H_2.append(ZXXI(), [[0], [1], [2], [3]])
H_2.append(IYYZ(), [[0], [1], [2], [3]])
H_2.append(IYYI(), [[0], [1], [2], [3]])
H_2.append(IXXZ(), [[0], [1], [2], [3]])
H_2.append(IXXI(), [[0], [1], [2], [3]])
```

```python
# define o divisor
divisor = QuantumCircuit(4, name=r'$U_{DF}=e^{i\pi/4\left(b^\dagger a+ba^\dagger\right)}$')
for i in range(0, j, 1):
    divisor.append(H_2, [[0], [1], [2], [3]])
    divisor.append(H_1, [[0], [1], [2], [3]])
divisor.draw('mpl', style='iqx')    
```

```python
# IBMQ.load_account()
provider = IBMQ.get_provider(
    hub='ibm-q', 
    group='open', 
    project='main'
)
simulator = Aer.get_backend('qasm_simulator')
nshots = 100000
```

```python
hom_j1 = QuantumCircuit(4, 4)

# preparação do estado de entrada do modo A, |1>_F = |01>
hom_j1.x(1) 
# preparação do estado de entrada do modo B, |1>_F = |01>
hom_j1.x(3) 
hom_j1.barrier()

# divisor de feixes
hom_j1.append(divisor, [[0], [1], [2], [3]])
hom_j1.barrier()

hom_j1.measure(0, 0)
hom_j1.measure(1, 1) 
hom_j1.measure(2, 2) 
hom_j1.measure(3, 3)
hom_j1.draw('mpl', style='iqx')
```

```python
plot_distribution(execute(hom_j1, backend = simulator, shots=nshots).result().get_counts(hom_j1))
```
